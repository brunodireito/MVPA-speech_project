```python
import pandas as pd
import csv
import os
```

```python
subj_map = {
   'AlexandreSayal': '0001',
}

task_map = {
   'run-inner_run1-data': 'TBD',
}

#ROOT_DATA="/Users/home/Documents/data/"
ROOT_DATA="/Volumes/LANGLYBACKU/DATA/DATA_Speech/"

TR=2
CONDITIONS=['Baseline ', 'Vogals', 'Syllables', 'Words', 'Phrases']
```

```python
sname='AlexandreSayal'
if sname in subj_map:
    print(subj_map[sname])
    sid=subj_map[sname]
    
    
tname='run-inner_run1-data'
if tname in task_map:
    print(task_map[tname])
    tcode=task_map[tname]
    
```

```python
# get .prt file
path=os.path.join(ROOT_DATA,sname,tname,'PROJECT','ANALYSIS')
prtfn = [f for f in os.listdir(path) if f.endswith('.prt')]

prtpath=os.path.join(path, prtfn[0])
print(prtpath)
```

```python
# read .prt file

events_num=[]

onset=[]
duration=[]
trial_type=[]

# Open file.
with open(prtpath, newline = '') as f_reader:
    # Read next line.
    l=f_reader.readline()
    # Iterate conditions.
    for idx, c in enumerate(CONDITIONS):
        
        while not c in l:
            l = f_reader.readline()
            if not l:
                break
        # read number of events.
        events_num.append(int(f_reader.readline()))
        print(c,int(events_num[idx]))
        # read events
        for e in range(int(events_num[idx])):
            temp_TR=f_reader.readline().split()
            # onset - TR to sec
            onset_sec=(int(temp_TR[0])-1)*TR       
            onset.append(onset_sec)
            duration_sec=(int(temp_TR[1])-int(temp_TR[0]))*TR
            # duration - TR to sec
            duration.append(duration_sec)
            trial_type.append(c)
```

```python
events = pd.DataFrame({'trial_type': trial_type,
                       'onset': onset,
                       'duration': duration})
```

```python
events
```

```python
events=events.sort_values('onset')
events=events.reset_index(drop=True)
events
```

```python
tsvfile = 'events_fonologico.tsv'
events.to_csv(tsvfile, sep='\t', index=False)
print("Created the events file in %s " % tsvfile)
```

```python
TR=2.5
onset_epoch=[]
duration_epoch=[]
trial_type_epoch=[]

for i in range(len(events)):
    idx_=((events['duration'][i]/TR))+1
    onset_=(events['onset'][i]) 
    for j in range(int(idx_)):   
        onset_epoch.append(onset_+(TR*j))
        duration_epoch.append(TR)
        trial_type_epoch.append(events['trial_type'][i])

events_epoch = pd.DataFrame({'trial_type': trial_type_epoch,
               'onset': onset_epoch,
               'duration': duration_epoch})
```

```python
tsvfile = 'events_epoch_fonologico.tsv'
events_epoch.to_csv(tsvfile, sep='\t', index=False)
print("Created the events file in %s " % tsvfile)
```
