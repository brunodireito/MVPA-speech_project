```python
import pandas as pd
import csv
import os
```

```python
subj_map = {
   'AndreiaVerdade': '0001',
   'BrunoDireito': '0002',
   'CarlosFerreira': '0003',
   'CarolinaTravassos': '0004',
   'CatarinaDuarte': '0005',
}

task_map = {
    'run-inner_run1-data': 'task-innerspeech_run-01',
    'run-inner_run2-data': 'task-innerspeech_run-02',
    'run-inner_run3-data': 'task-innerspeech_run-03',
}

#ROOT_DATA="/Users/home/Documents/data/"
#ROOT_DATA="/Volumes/LANGLYBACKU/DATA/DATA_Speech/"
ROOT_DATA="/media/brunomiguel/TOSHIBA EXT/DATA_Carlos_PhD_Final"

TR=2
CONDITIONS=['Baseline','Vogals','Syllables','Words','Phrases']
```

```python
sname='AndreiaVerdade'
if sname in subj_map:
    print(subj_map[sname])
    sid=subj_map[sname]
    
    
tname='run-inner_run2-data'
if tname in task_map:
    print(task_map[tname])
    tcode=task_map[tname]
    
```

```python
# get .prt file
path=os.path.join(ROOT_DATA,sname,tname,'PROJECT','ANALYSIS')
prtfn = [f for f in os.listdir(path) if f.endswith('.prt')]

prtpath=os.path.join(path, prtfn[0])
print(prtpath)
```

```python
# read .prt file

events_num=[]

onset=[]
duration=[]
trial_type=[]

# Open file.
with open(prtpath, newline = '') as f_reader:
    # Read next line.
    l=f_reader.readline()
    # Iterate conditions.
    for idx, c in enumerate(CONDITIONS):
        
        while not c in l:
            l = f_reader.readline()
            if not l:
                break
                
        # Read number of events of type c.
        events_num.append(int(f_reader.readline()))
        print('Condition - ' + c + ', number of repetitions - ' + str(events_num[idx]))
        
        # Iterate through events of type c.
        for e in range(int(events_num[idx])):
            temp_TR=f_reader.readline().split()
            
            # onset - TR to sec.
            onset_sec=(int(temp_TR[0])-1)*TR       
            onset.append(onset_sec)
            
            # duration - TR to sec
            duration_sec=(int(temp_TR[1])-(int(temp_TR[0])-1))*TR
            duration.append(duration_sec)
            trial_type.append(c)
```

```python
events = pd.DataFrame({'trial_type': trial_type,
                       'onset': onset,
                       'duration': duration})
```
