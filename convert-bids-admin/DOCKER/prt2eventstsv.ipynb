```python
# read .prt file
root_data="E:\\DATA\\DATA_Carlos_Phonos\\"
root_data="/Users/home/Documents/data/"

prt_file=root_data+'JoaoPereira/001/run-fonologico-data/PROJECT/ANALYSIS/Fonologico.prt'

conds=['Baseline ', 'Task ']
c_numel=[]
events_num=[]
TR=2.5
onset=[]
duration=[]
trial_type=[]

# open file.
with open(prt_file, newline = '') as f_reader:
    l=f_reader.readline()
    for idx, c in enumerate(conds):
        while not c in l:
            l = f_reader.readline()
        # read number of events.
        events_num.append(int(f_reader.readline()))
        print(c,int(events_num[idx]))
        # read events
        for e in range(int(events_num[idx])):
            temp_TR=f_reader.readline().split()
            # onset - TR to sec
            onset_sec=(int(temp_TR[0])-1)*TR       
            onset.append(onset_sec)
            duration_sec=(int(temp_TR[1])-int(temp_TR[0]))*TR
            # duration - TR to sec
            duration.append(duration_sec)
            trial_type.append(c)
```

```python
import pandas as pd
events = pd.DataFrame({'trial_type': trial_type,
                       'onset': onset,
                       'duration': duration})
```

```python
events
```

```python
events=events.sort_values('onset')
events=events.reset_index(drop=True)
events

```

```python
tsvfile = 'events_fonologico.tsv'
events.to_csv(tsvfile, sep='\t', index=False)
print("Created the events file in %s " % tsvfile)
```

```python
TR=2.5
onset_epoch=[]
duration_epoch=[]
trial_type_epoch=[]

for i in range(len(events)):
    idx_=((events['duration'][i]/TR))+1
    onset_=(events['onset'][i]) 
    for j in range(int(idx_)):   
        onset_epoch.append(onset_+(TR*j))
        duration_epoch.append(TR)
        trial_type_epoch.append(events['trial_type'][i])

events_epoch = pd.DataFrame({'trial_type': trial_type_epoch,
               'onset': onset_epoch,
               'duration': duration_epoch})
```

```python
tsvfile = 'events_epoch_fonologico.tsv'
events_epoch.to_csv(tsvfile, sep='\t', index=False)
print("Created the events file in %s " % tsvfile)
```

```python
TR=2.5
num_events=222
betaseries_df=[]
onset_betaseries=[]
duration_betaseries=[]
trial_type_betaseries=[]

for i in range(num_events):
    onset_betaseries.append(TR*i)
    duration_betaseries.append(TR)
    trial_type_betaseries.append(i)

betaseries_df = pd.DataFrame({'trial_type': trial_type_betaseries,
               'onset': onset_betaseries,
               'duration': duration_betaseries})
```
